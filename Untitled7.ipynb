```
#---------------Attention Span Min Max---------------
media_min_max = media.groupby(['contact_id', 'video_id']).agg({'start_position':'min', 'end_position':'max'}).reset_index()
media = media.merge(media_min_max, on=['contact_id','video_id'], how='left')


```

```
#---------Attention Span Min Max---------------
# Attention Span for a video watched multiple times by customer


media['Attention_Span_min_max'] = (media['end_position'] - media['start_position'])/media['Length']

media.head()
```

```
media = media.dropna(axis=0, subset=['Attention_Span_min_max'])
media.head()
```

```
#-----------Attention Score---------------------------
conditions = [
    (media['Attention_Span_min_max'] >= 0) & (media['Attention_Span_min_max'] <= 0.25),
    (media['Attention_Span_min_max'] > 0.25) & (media['Attention_Span_min_max'] <= 0.50),
    (media['Attention_Span_min_max'] > 0.5) & (media['Attention_Span_min_max']<=0.75),
    (media['Attention_Span_min_max'] > 0.75)
    ]

# create a list of the values we want to assign for each condition
values = [5, 10, 15, 20]

# create a new column and use np.select to assign values to it using our lists as arguments
media['Attention_Score'] = np.select(conditions, values)

# display updated DataFrame
media.head()
```

```
#Number of Replays:

media['Final_Score'] = media['Attention_Score'] + (media['Finishes']*25)

media.head()
```
